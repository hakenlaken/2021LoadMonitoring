```python
db = pd.read_csv('/content/data_2_test.csv')
db
```

```python
db.columns = ['data','pi','cpu','ram']
db.head(20)
```

```python
db.drop(db.index[:3], inplace=True)
db.head(20)
```

```python
db['delta_pi'] = 0
db['delta_pi'] = db['pi'].diff()
db = db.iloc[1:]
db.head(20)
```

```python
db.head(10)
```

```python
pd.options.mode.chained_assignment = None
db['cpu'] = list(map(lambda x: x[:-1], db['cpu'].values))
db['ram'] = list(map(lambda x: x[:-1], db['ram'].values))
```

```python
db
```

```python
db['cpu'] = [float(x) for x in db['cpu'].values]
db['ram'] = [float(x) for x in db['ram'].values]
```

```python
db.head(30)
```

```python
print(db['cpu'].corr(db['delta_pi']))
print(db['ram'].corr(db['delta_pi']))
```

```python
db.head(30)
db.reset_index(inplace=True)
```

```python
db["data"] = pd.to_datetime(db['data'], format='%Y-%m-%dT%H:%M:%S.%f')
db['TimeStamp'] = 1

```

```python
past_seconds = 5

#for second in range(past_seconds):
#  s = second+1
#  db[f"delta_pi_back_{s}s"] = db["delta_pi"].shift(s)

db["delta_pi_5s_mean"] = db["delta_pi"].shift(1).rolling(window=5).median()
db["delta_pi_15s_mean"] = db["delta_pi"].shift(1).rolling(window=15).median()
db["delta_pi_25s_mean"] = db["delta_pi"].shift(1).rolling(window=25).median()
db["delta_pi_35s_mean"] = db["delta_pi"].shift(1).rolling(window=35).median()
db["delta_pi_45s_mean"] = db["delta_pi"].shift(1).rolling(window=45).median()
db["delta_pi_55s_mean"] = db["delta_pi"].shift(1).rolling(window=55).median()
db["delta_pi_65s_mean"] = db["delta_pi"].shift(1).rolling(window=65).median()
```

```python

db.dropna(inplace=True)
db.head(50)
```

```python
db.reset_index(inplace=True)
db.head(30)
```

```python
df = db
```

```python
df
```

```python
df.drop(['level_0','data','index','cpu','ram','pi'], axis=1, inplace=True)
df.head(30)
```

```python
X = df # Входные данные
Y = df["delta_pi"] # Выход
```
