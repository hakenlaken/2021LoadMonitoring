```python
db = pd.read_csv('/content/data_2_test.csv')
db
```

```python
db.columns = ['data','pi','cpu','ram']
db.head(20)
```

```python
db.drop(db.index[:3], inplace=True)
db.head(20)
```

```python
db['delta_pi'] = 0
db['delta_pi'] = db['pi'].diff()
db = db.iloc[1:]
db.head(20)
```

```python
db.head(10)
```

```python
pd.options.mode.chained_assignment = None
db['cpu'] = list(map(lambda x: x[:-1], db['cpu'].values))
db['ram'] = list(map(lambda x: x[:-1], db['ram'].values))
```

```python
db
```

```python
db['cpu'] = [float(x) for x in db['cpu'].values]
db['ram'] = [float(x) for x in db['ram'].values]
```

```python
db.head(30)
```

```python
print(db['cpu'].corr(db['delta_pi']))
print(db['ram'].corr(db['delta_pi']))
```
